```python
# Bài 2.1. Cho dữ liệu có chứa chỉ số chiều cao và cân nặng của 200 người được lưu trong tệp
# height_weight.csv
# a) Viết chương trình sử dụng thuật toán K-mean để phân cụm dữ liệu trên thành k
# cụm (k = 2, 3, 4, 5)
# b) Trực quan hóa các cụm dữ liệu ứng với các k ở câu 2.1.a)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('height_weight.csv')
df.tail(5)
```

```python
# a) Viết chương trình sử dụng thuật toán K-mean để phân cụm dữ liệu trên thành kcụm (k = 2, 3, 4, 5)
# Tạo một danh sách lưu lại các nhãn
cluster_labels={}
X=df.values
# Áp dụng hàm phân cụm có dùng thư viện
for k in [2,3,4,5]:
    Kmeans=KMeans(n_clusters=k,random_state=42)
    labels=Kmeans.fit_predict(X)
    cluster_labels[k]=labels
    print(f"K={k}: \nTậm cụm: \n{Kmeans.cluster_centers_} \nNhãn cụm: \n{labels}\n")
```

```python
# b) Trực quan hóa các cụm dữ liệu ứng với các k ở câu 2.1.a)
# Hàm thực hiện phân cụm và trực quan hóa
def cluster_and_visualize(X, k_values):
    for k in k_values:
        # Áp dụng K-means
        kmeans = KMeans(n_clusters=k, random_state=42)
        y_kmeans = kmeans.fit_predict(X)
        
        # Trực quan hóa cụm
        plt.figure(figsize=(8, 6))
        for i in range(k):
            plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], label=f'Cụm {i+1}')
        
        # Tâm cụm
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                    s=200, c='red', marker='X', label='Tâm cụm')
        plt.title(f'Phân cụm K-means với k = {k}')
        plt.xlabel('Chiều cao')
        plt.ylabel('Cân nặng')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
X=df.values
cluster_and_visualize(X, k_values=[2, 3, 4, 5])
```
