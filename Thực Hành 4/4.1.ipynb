Cho tập dữ liệu huấn luyện ước lượng khả năng mua các dòng xe theo các độ tuổi như
sau:    

UserID Age Car Risk
1. 25 Sports L
2. 20 Vintage H
3. 25 Sports L
4. 45 SUV H
5. 20 Sports H
6. 25 SUV H

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'UserID': [1, 2, 3, 4, 5, 6],
    'Age': [25, 20, 25, 45, 20, 25],
    'Car': ['Sports', 'Vintage', 'Sports', 'SUV', 'Sports', 'SUV'],
    'Risk': ['L', 'H', 'L', 'H', 'H', 'H']
}
```

```python
df = pd.DataFrame(data)
df
```

* UserID: mã định danh, không quan trọng
* Age: tuổi 
* car: xe/ loại xe (có 3 loại là Sports- thể thao; Vintage- cổ điển; SUV- ???)
* Risk: rủi ro mua xe 

Chuyển đổi 

```python
label_encoder_car = LabelEncoder()
df['Car'] = label_encoder_car.fit_transform(df['Car'])

label_encoder_risk = LabelEncoder()
df['Risk'] = label_encoder_risk.fit_transform(df['Risk'])
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

Tách dữ liệu

```python
x=df[['Age','Car']]
y=df['Risk']
```

Huấn luyện mô hình

```python
clf=DecisionTreeClassifier()
clf.fit(x,y)
```

Kiểm tra với dữ liệu mới

```python
new_data = pd.DataFrame({
    'Age': [25, 23],
    'Car': label_encoder_car.transform(['Sports', 'SUV'])
})
```

Đưa vào kiểm tra

```python
predictions = clf.predict(new_data)
```

Giải mã và hiển thị kết quả

```python
# Giải mã
predicted_labels = label_encoder_risk.inverse_transform(predictions)

# Hiển thị kết quả
test_data_with_predictions = new_data.copy()
test_data_with_predictions['Predicted Risk'] = predicted_labels
test_data_with_predictions['Car'] = label_encoder_car.inverse_transform(new_data['Car'])

print(test_data_with_predictions)
```

## Kết luận
25-Sports : thuộc nhãn L-low Risk; Tỷ lệ mua thấp   
23-SUV    : thuộc nhãn H-high Risk; Tỷ lệ mua cao
